An ever growing body of scientific research supports the theory that greenhouse gases emitted by human activity are [dangerously warming Earth's atomosphere.](https://climate.nasa.gov/evidence/) Modern transportation, often achieved by burning fossil fuels, is a significant contributor of these greenhouse gases, namely carbon dioxide (CO<sub>2</sub>). Of transportation emissions, [61% come from privately owned passenger cars and trucks,](https://www.epa.gov/greenvehicles/fast-facts-transportation-greenhouse-gas-emissions) which amounts to 16.5% of all American emissions.  

Given the United States' [large emission footprint globally](https://www.epa.gov/ghgemissions/global-greenhouse-gas-emissions-data) and the significant contribution of greenhouse gasses from private vehicles, the government is acting to reduce greenhouse gas emissions from cars and trucks sold in America. Under direction from President Obama in 2010, the EPA and the National Highway Traffic Safety Administration (NHTSA) [jointly developed new fuel economy standards](https://www.nhtsa.gov/laws-regulations/corporate-average-fuel-economy#9491) which car companies must meet in order to sell their vehicles. You can read more about the history, goals, and effectiveness of fuel economy standards in the United States and Europe [here](http://www.rff.org/files/document/file/RFF-PB-16-03.pdf), or about the Trump administration's efforts to review the latest standards [here](http://www.nbcnews.com/business/autos/trump-rolls-back-obama-era-fuel-economy-standards-n734256).  

With fuel economy standards increasing over time, it stands to reason to that cars and trucks should be getting more fuel efficient over time as well. Beyond that, simple economics suggests that consumers should prefer to purchase more fuel efficient vehicles. Why get a car with 15mpg that's expensive to drive when you can get one with 30mpg, right? Thus, in this tutorial we explore the fuel efficiency of cars from 1984 to 2017. We obtained a CSV file with fuel economy measures for over 37,000 vehicle models sold in the United States since 1984 from [Kapsarc](https://datasource.kapsarc.org/page/about/), an energy focused data science portal which last scraped the data on 4/30/17 from [fueleconomy.gov](https://www.fueleconomy.gov/feg/ws/index.shtml), an EPA hosted site. How has the average highway mpg of vehicles changed over time compared to city mpg? Did the 2010 fuel economy standards signficantly change the rate of improvement in vehicle fuel efficiency? We'll aim to answer these and other questions while we explore this [dataset](https://datasource.kapsarc.org/explore/dataset/us-vehicle-fuel-economy-data-1984-2017/table/).  

## I: Importing and Cleaning the Dataset

```python
#Let's first import some useful python libraries
import pandas as pd
import seaborn as sns
import numpy as np
import re
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import decimal
```

```python
#Next, let's import the dataset and take a quick view at its contents
df = pd.read_csv("us-vehicle-fuel-economy-data-1984-2017.csv", sep=';')
pd.set_option('display.max_columns', None)
print(df.shape)
df.head()
```

If you didn't check out the dataset link in the introduction, the full description of the dataset and every column's meaning can be found [here.](https://datasource.kapsarc.org/explore/dataset/us-vehicle-fuel-economy-data-1984-2017/information/) There are 37,129 vehicles with 83 attributes in the dataset. Looking over the .head() output, the columns of the dataset appear to be arranged in no particular order. Scrolling through them suggests that each row corresponds to a unique vehicle model via the primary key "id", which roughly corresponds to a unique combination of the attributes "make", "model", "year," and "trany" (indicating transmission type.) The first thing we'll do then is reorganize the columns so these attributes appear at the beginning of the dataset and so the rest follow alphabetically.

```python
#resort the columns by identifying information at the front, then alphabetically
head = ['id', 'make', 'model', 'year', 'trany']
cols = [x for x in sorted((df)) if x not in head]
cols = head + cols
df = df[cols].sort_values(by='id')

#view the newly organized dataset
df.head()
```

That's nicer! Still, we have a very large number of attributes, and it also appears that many of the columns of the dataset are mostly empty with values of 0, -1, or NaN. We don't believe trying to impute missing values for such mostly empty attributes would be appropriate given our lacking of understanding of the distribution from which the measurements are drawn. Some attributes are also redundant in the information they record. "barrels08" for instance logs the annual petroleum consumption in barrels of a car, but this is a direct function of its "comb08" (combined city/hwy mpg) attribute. Another example is "UCity" and "UHighway", which are [outdated mpg estimates](https://www.fueleconomy.gov/feg/ratings.shtml) compared to "city08" and "highway08." Other attributes such as "createdOn" record information about when the vehicle was added to the dataset which isn't a useful attribute for our analysis.

For these reasons, we'd like to reduce the number of attributes down to those we are interested in, those which aren't redundant, and those which are not mostly empty values for the entire dataset. We thus remove the attributes listed below, as well as any attributes we're not interested in for which less than 1% of cars have a non-empty value of the attrbute.

```python
#Redundant or non-useful attributes we'll remove are listed here.
#We remove fuelCost attributes on the basis that fuel prices change over time,
#so these attributes only encapsulate the fuelCost at a given time.
#We also remove all EPA emissions scores as they are a direct function of a car's
#fuelType and co2 tailpipe output.
drop = ['UCity', 'UCityA', 'atvType', 'barrels08', 'barrelsA08', 'fuelType', \
       'UHighway', 'UHighwayA', 'createdOn', 'modifiedOn', 'cityA08U', 'city08U', \
       'co2', 'co2A', 'comb08U', 'combA08U', 'engId', 'eng_dscr', 'fuelCost08', \
       'fuelCostA08', 'guzzler', 'highway08U', 'highwayA08U', 'mfrCode', \
       'mpgData', 'trans_dscr', 'VClass', 'youSaveSpend', 'hlv', 'hpv', 'lv2', \
       'lv4', 'pv2', 'pv4', 'ghgScore', 'ghgScoreA', 'feScore', 'rangeA']

#Collect columns which are are mostly empty.
#The print statement can be uncommented to see which
#rows have empty values and by what percentage.
empty_attrs = []
empty_vals = [np.NaN, 0, -1]
for col_name in cols:
    col = df[col_name]
    non_empty = [x for x in col if x not in empty_vals]
    #print(col_name, len(non_empty) / len(col))
    if len(non_empty) / len(col) < 0.01:
        empty_attrs.append(col_name)
        
#remove unwanted attributes
drop = list(set().union(empty_attrs, drop))
df = df.drop(drop, axis = 1)
df.head()
```

Great! Our last data tidying steps will be to create some dummy variables for categorical attributes in the dataset such as "trany", "drive", and "fuelType1" so we can use these attributes in later analyses, to impute some missing measures for some attributes like "displ" (cylinder capacity in liters) and "cylinders", and to clean the evMotor attribute. 

```python
#This commented out 3 lines below can be used to print out the range of values for the 
#categorical variables we wanted to create dummies for and was used to help
#write the dummy creator code below it.

#cats = ['trany', 'drive', 'sCharger', 'tCharger', 'startStop', 'fuelType1', 'fuelType2']
#for cat in cats:
#    print(cat, df[cat].unique())

#create our dummy vars for each row of data and impute some values
for ix, row in df.iterrows():
    trans = row['trany']
    drive = row['drive']
    f1 = row['fuelType1']
    f2 = row['fuelType2']
    sCharge = row['sCharger']
    tCharge = row['tCharger']
    ss = row['startStop']
    evM = row['evMotor']
    cyl = row['cylinders']
    displ = row['displ']
     
    #transmission will be split into 2 attributes: the dummy automatic? , for which 0
    #means the car is manual and 1 means the car is automatic, and the numeric
    #attribute gears. For transmissions with no recorded number of gears, we
    #sample from the data for that year to impute a value.
    if not pd.isnull(trans):
        df.set_value(ix, 'automatic?', 1 if 'Auto' in trans else 0)
        gears = re.findall(r'\d+', trans)
        if len(gears) > 0:
            df.set_value(ix, 'gears', gears[0])
        else:
            gears = np.NaN
            while pd.isnull(gears):
                gears = list(df.loc[(df['year'] == row['year'])]['gears'].sample(n=1))[0]
            df.set_value(ix, 'gears', gears)
    #A handful of vehicles have no recorded transmission. We are not sure why, but we treat 
    #these records as missing at random and impute them by sampling from the distribution of
    #values from the year of data in question.
    else:
        auto = np.NaN
        while pd.isnull(auto):
            auto = list(df.loc[(df['year'] == row['year'])]['automatic?'].sample(n=1))[0]
        df.set_value(ix, 'automatic?', 0 if auto == 1 else 1)
        gears = np.NaN
        while pd.isnull(gears):
            gears = list(df.loc[(df['year'] == row['year'])]['gears'].sample(n=1))[0]
        df.set_value(ix, 'gears', gears)
    
    
    #next we'll tidy up drive. drive is recorded inconsistently, sometimes as front wheel,
    #rear wheel, or all wheel, or sometimes simply as 2 wheel or 4 wheel. We simplify
    #the attribute to the simple dummies 4-wheel?, for which 1 means 4 wheel and 0 means 
    #2 wheel drive, and rear?, for which 1 means rear wheel drive and 0 means front wheel drive
    if not pd.isnull(drive):
        df.set_value(ix, '4WheelDrive?', 1 if 'All' in drive or '4' in drive else 0)
        df.set_value(ix, 'rear?', 1 if 'Rear' in drive else 0)
    #As was the case for transmission, some vehicles are missing a drive variables. This accounts
    #for nearly 4% of the data. Instead of trying to impute values for so much data for which
    #we cannot discern the reason for missingness, we set their dummy attributes to 0.
    else:
        df.set_value(ix, '4WheelDrive?', 0)
        df.set_value(ix, 'rear?', 0)
        
        
    #next are the the fuelType vars. We'll split fuelType1 into dummies based on type
    #and create a single fuelType2? dummy from fuelType2. A car with 0 for all
    #of this dummies indicates a car with uses regular gas
    df.set_value(ix, 'premiumGas?', 1 if 'Premium' in f1 else 0)
    df.set_value(ix, 'midgradeGas?', 1 if 'Midgrade' in f1 else 0)
    df.set_value(ix, 'diesel?', 1 if 'Diesel' in f1 else 0)
    df.set_value(ix, 'naturalGas?', 1 if 'Natural' in f1 else 0)
    df.set_value(ix, 'electric?', 1 if 'Electric' in f1 else 0)
    df.set_value(ix, 'fuelType2?', 1 if type(f2) == str else 0)
    
    #next dummies for the turbo and super charged and startStop attributes
    df.set_value(ix, 'tCharger', 1 if type(tCharge) == str else 0)
    df.set_value(ix, 'sCharger', 1 if type(sCharge) == str else 0)
    df.set_value(ix, 'startStop', 1 if type(ss) == str and ss == 'Y' else 0)
    
    
    #That's it for the dummy variables. Now we'll clean evMotor which has NaN's
    #and non standard syntax instead of a simple int value
    if not pd.isnull(evM):
        kw_hrs = re.findall(r'\d+', evM)
        df.set_value(ix, 'kwHrs', kw_hrs[0])
    else:
        df.set_value(ix, 'kwHrs', 0)
    
    #Finally, we impute displ and cylinders for a handful of cars missing this data. We recognize 
    #electric cars have no cylinders or displ, so we impute 0 for electric cars missing this data.
    #We treat other missing records as missing at random and sample from the dataset to impute
    #these values.
    if pd.isnull(displ):
        if 'Electric' in f1:
            df.set_value(ix, 'displ', 0)
        else:
            while pd.isnull(displ):
                displ = list(df.loc[(df['year'] == row['year'])]['displ'].sample(n=1))[0]
            df.set_value(ix, 'displ', displ)
    if pd.isnull(cyl):
        if 'Electric' in f1:
            df.set_value(ix, 'cylinders', 0)
        else:
            while pd.isnull(cyl):
                cyl = list(df.loc[(df['year'] == row['year'])]['cylinders'].sample(n=1))[0]
            df.set_value(ix, 'cylinders', cyl)
    
    
#Finally delete our categorical attributes we no longer need and sort the attributes as before
drop = ['fuelType1', 'drive', 'trany', 'fuelType2', 'evMotor']
df = df.drop(drop, axis = 1)
head = ['id', 'make', 'model', 'year']
cols = [x for x in sorted((df)) if x not in head]
cols = head + cols
df = df[cols].sort_values(by='id')
df.head()
```

Excellent. Now we're ready to explore the dataset.

## II: Exploring Vehicle Fuel Economy Over Time

We'll start out by looking at the fuel efficiency of cars sold in America over time as well as the amount of greenhouse gases cars are giving off. The primary measure of a vehicle's greenhosue gas output is grams of CO<sub>2</sub> output per mile of driving from the tailpipe since [CO<sub>2</sub> makes up 96% of greenhouse gasses emitted by the average vehicle](https://nepis.epa.gov/Exe/ZyPDF.cgi?Dockey=P100ONBL.pdf). To visualize the distributions of vehicle fuel efficiencies and CO<sub>2</sub> output over time, we'll use violin plots. This type of plot is similar to a box plot in that for each year, it shows the median of a variable's distribution (labeled by a dot in the center) and the interquartile range surrounding it, but additionally allows visualization of the distribution's probability density. This plot component makes skews in each distribution slightly clearer and also points out when a distribution is multimodal.

Below, we generate violin plots for how highway gas mileage, city gas mileage, and CO<sub>2</sub> emission readings have changed in vehicles models over time.

```python
sns.set_style("whitegrid")

#create a violin plot for vehicle highway mpg vs. time
plt.figure(figsize=(12,6))
ax = sns.violinplot(x="year", y="highway08", data=df)
plt.title('Vehicle Highway Gas Mileage vs. Time', fontsize=20)
plt.setp(ax.get_xticklabels(), rotation=45, fontsize=12)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Highway MPG', fontsize=14)
plt.show()

#create a violin plot for vehicle city mpg vs. time
plt.figure(figsize=(12,6))
ax = sns.violinplot(x="year", y="city08", data=df)
plt.title('Vehicle City Gas Mileage vs. Time', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('City MPG', fontsize=14)
plt.setp(ax.get_xticklabels(), rotation=45, fontsize=12)
plt.show()

#create a violin plot for vehicle CO2 g/mi vs. time
plt.figure(figsize=(12,6))
ax = sns.violinplot(x="year", y="co2TailpipeGpm", data=df)
plt.title('Vehicle CO2 Emissions vs. Time', fontsize=20)
plt.setp(ax.get_xticklabels(), rotation=45, fontsize=12)
plt.xlabel('Year', fontsize=16)
plt.ylabel('CO2 Grams/Mile', fontsize=14)
plt.show()
```

Looking first at the plots for gas mileage over time, we see somewhat wide distributions in both highway mpg and city mpg measures each year. Many distributions have noticeably high right skews, indicating there are some vehicles each year with much higher mpg measures than the majority of other cars. Turning to the CO<sub>2</sub> emissions vs. time plot, we also see wide distributions in each year, with earlier years having very high right skews indicating that some cars with particularly high CO<sub>2</sub> emissions were manufactured at that time. Conversely, some later years have noticeably low distribution tails. As we'll see in the next section, the outliers causing the left skew in this plot and the right skew in the mileage plots are electric powered vehicles.

Next, looking closely at the medians in each plot (indicated by white dots in each distribution), we notice that they slightly increase over time in the gas mileage plots and decrease slightly over time for the CO<sub>2</sub> emissions one. However, because of y-axis scale, this relatively modest slope is hard to see. 

We can better examine the change in center for each of these variables by plotting the means for each year in scatter plots. While the mean is still an imperfect metric for central tendency in this dataset given the extreme outliers we see, we chose it for a scatter plot because it yields non-discrete values, better highlighting differences between consecutive years.

```python
#we'll group cars by year, calculate mean measure for the variables we're inspecting
#for each year, and put these in a new dataframe to graph each variable's mean vs time
groups = df.groupby(['year'])
group_agg = groups.aggregate(np.mean)
collapsed_group_agg = pd.DataFrame(columns = ['year', 'meanHighway', 'meanCity', 'meanCO2'])
for i in range(len(group_agg)):
    row = {
        'year': group_agg.iloc[i].name,
        'meanHighway': group_agg.iloc[i].highway08,
        'meanCity': group_agg.iloc[i].city08,
        'meanCO2': group_agg.iloc[i].co2TailpipeGpm
    }
    collapsed_group_agg = collapsed_group_agg.append(row, ignore_index=True)

#create scatterplot for mean vehicle highway mpg vs. time
plt.figure(figsize=(12,6))
plt.title("Mean Highway MPG vs Time", fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('MPG', fontsize=14)
plt.scatter(collapsed_group_agg['year'], collapsed_group_agg['meanHighway'])
plt.show()
 
#create scatterplot for mean vehicle city mpg vs. time
plt.figure(figsize=(12,6))
plt.title("Mean City MPG vs Time", fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('MPG', fontsize=14)
plt.scatter(collapsed_group_agg['year'], collapsed_group_agg['meanCity'])
plt.show()

#create scatterplot for mean vehicle CO2 g/mi vs. time
plt.figure(figsize=(12,6))
plt.title("Mean CO2 Emissions vs Time", fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('CO2 Tailpipe GPM', fontsize=14)
plt.scatter(collapsed_group_agg['year'], collapsed_group_agg['meanCO2'])
plt.show()
```

Looking first at the gas mileage plots, it appears that the average car's highway and city gas mileage both didn't change much at all between 1984 and 2007. The average car's highway mpg hovered near 23mpg during that time period, while the mean city mpg varied between 16 and 18mpg. From 2008 onward however, we see both measures begin to increase roughly linearly from year to year. Looking at the overall increases between 2008 and 2016, we also see that mean highway gas mileage increases a bit faster than mean city gas mileage each year on average.

Next analyzing the CO<sub>2</sub> emissions plot, its shape is almost a mirror opposite of both gas mileage plots. While the mean CO<sub>2</sub> grams emitted per mile fluctuates around 490grams/mile from 1984 to 2007, afterwards it begins to decrease linearly each year. This inverse relationship between gas mileage and CO<sub>2</sub> emissions is indicative that CO<sub>2</sub> grams emitted per mile is very much a function of a car's gas mileage.

## III: Exploring the Rise of Electric and Hybrid Vehicles

We'll also explore trends in the growth of electric and hybrid vehicles. Let's first produce plots for the number of hybrid vehicle models and electric vehicle models vs. time.

```python
#create a simple data frame with year and model count for hybrid cars
hybrids = df.loc[(df['fuelType2?'] == 1)]
hybridAgg = pd.DataFrame(columns = ['Year', 'Vehicle Models'])
sums = dict(hybrids.groupby('year').size())
for year, count in sums.items():
    hybridAgg = hybridAgg.append([{'Year': year, 'Vehicle Models': count}])

#create a simple data frame with year and model count for electric cars
electrics = df.loc[(df['electric?'] == 1)]
electricAgg = pd.DataFrame(columns = ['Year', 'Vehicle Models'])
sums = dict(electrics.groupby('year').size())
for year, count in sums.items():
    electricAgg = electricAgg.append([{'Year': year, 'Vehicle Models': count}])

#plot each data frame as a stripplot
plt.figure(figsize=(12,6))
hybridPlot = sns.stripplot(x='Year', y='Vehicle Models', data=hybridAgg)
plt.title('Growth of Hybrid Vehicles over Time', fontsize=20)
hybridPlot.set_xticklabels(hybridAgg['Year'].astype(int), rotation=35)
sns.plt.show()
plt.figure(figsize=(12,6))
electricPlot = sns.stripplot(x='Year', y='Vehicle Models', data=electricAgg)
plt.title('Growth of Electric Vehicles over Time', fontsize=20)
electricPlot.set_xticklabels(electricAgg['Year'].astype(int), rotation=35)
sns.plt.show()
```

We see that hybrid vehicles were first introduced to the American market in 2000, with electric cars being introduced two years prior. The number of hybrid models appears to have grown steadily ever since their introduction. There's a bit of a dropoff in the number of models offered in 2015 and 2016 compared to prior years, but this may be because not all vehicles made in those years have been added to the dataset yet. As for electric vehicle models, after some intial growth from 1998-2003, only one model was produced between 2003 and 2010 before production took off again in 2011. An interesting trend that can be seen in both plots is a spike in growth for the number of models offered each year from 2011 onward, which may be a reaction by car manufacturers to the EPA's 2010 fuel economy regulations.

Let's next see how hybrid vehicle technology has improved over time by creating violin plots for the city and highway fuel economies of such vehicles and the CO<sub>2</sub> grams output per mile by hybrid vehicles vs. the year.

```python
#create violin plot for highway08 vs. year for hybrid cars
plt.figure(figsize=(12,6))
hybridPlot = sns.violinplot(x='year', y='highway08', data=hybrids)
plt.title('Hybrid Vehicle Highway Gas Mileage vs. Time', fontsize=20)
hybridPlot.set_xticklabels(hybrids['year'].unique().astype(int), rotation=35)
hybridPlot.set(xlabel='Year', ylabel='Highway MPG')
sns.plt.show()

#create violin plot for city08 vs. year for hybrid cars
plt.figure(figsize=(12,6))
hybridPlot = sns.violinplot(x='year', y='city08', data=hybrids)
plt.title('Hybrid Vehicle City Gas Mileage vs. Time', fontsize=20)
hybridPlot.set_xticklabels(hybrids['year'].unique().astype(int), rotation=35)
hybridPlot.set(xlabel='Year', ylabel='City MPG')
sns.plt.show()

#create violin plot for co2TailpipeAGpm vs. year for hybrid cars
plt.figure(figsize=(12,6))
hybridPlot = sns.violinplot(x='year', y='co2TailpipeAGpm', data=hybrids)
plt.title('Hybrid Vehicle CO2 Emissions vs. Time', fontsize=20)
hybridPlot.set_xticklabels(hybrids['year'].unique().astype(int), rotation=35)
hybridPlot.set(xlabel='Year', ylabel='CO2 Grams/Mile')
sns.plt.show()
```

Overall though, we can see that the latest electric cars are roughly 4 times more energy efficient than the average car sold in America. The latest electric cars achieve nearly 100MPGe in cities and on the highway, while our earlier plots suggest that the latest average car gets just above 20mpg in the city and nearly 29mpg on highways.

Our next plot shows that battery technology has began improving for electric vehicles since 2011 after relative stagnation from 1998-2002. From 50kWh in 2011, the median battery capcity of electric vehicles doubled to 100kWh by 2014 and nearly doubled again by 2016. The final plot shows that every electric vehicle in the dataset is measured as giving off no CO<sub>2</sub> while being driven (0 grams/mi). The title "zero carbon footprint" should be taken with a grain of salt however, as the type of energy that produces the electricity to charge these cars likely produces [upstream emissions](https://www.fueleconomy.gov/feg/evghg.shtml) depending on where in the country each car is driven.

## IV: What Determines a Car's Fuel Economy?

Now, let's look at how different physical vehicle attributes affect a vehicle's gas mileage by doing multilinear regressions.

Taking our initial set of variables, we first will see what affects a car's city mpg (city08). That would make the remaining variables our predictors. However, upon closer inspection, many of these variables are intuitively not useful to keep in our model. Variables "id", "make" and "model" are simply categorical variables that would not be helpful in predicting a car's mpg. We also removed comb08, and highway08 which are also measures of mpg of the car and therefore directly correlated with city08. Another variable to remove is co2TailpipeGpm, which as we saw earlier appears to be dependent upon a car's gas mileage. Finally, cityA08, highwayA08, combA08, and co2TailpipeAGpm were all removed because these represent different metrics for fuel efficiency for the alternative fuel types that a car may run on. Because many cars do not have an alternative fuel type, it made sense to exclude these from predicting mpg for a regression on all the cars in our dataset. So first we'll drop these variables.

```python
#creating a linear regression model to predict city08

#remove unrelated predictors
predictors = df.drop(['year','id', 'make', 'model', 'city08', 'cityA08', 'co2TailpipeAGpm', \
                      'co2TailpipeGpm', 'comb08', 'combA08', 'highway08', 'highwayA08'], \
                     axis = 1).astype(float)

#add a constant term to our regression
x = sm.add_constant(predictors.values)

#convert to np.array to prepare for regression
y = df['city08'].as_matrix()
```

Now that we've done some initial processing of the dataset, let's next see if any of the remaining predictive variables need to be [eliminated due to multicollinearity.](https://onlinecourses.science.psu.edu/stat501/node/347) Such variables complicate and overfit a model, so before we perform the regression, we'll compute the variance inflation factor(VIF) of each variable to detect multicollinearity.

```python
#first let's see if our model has high multicollinearity
vif = [variance_inflation_factor(x, ix) for ix in range(x.shape[1])]
print("These are the VIF scores of each predictor:")
print (vif)
```

Based off of this output, it seems that we have some highly correlated variables, which are represented by VIF scores greater than 5. Thus, let's go ahead and remove such predictors from the dataset until we have no more highly correlated variables.

```python
#This function drops the variable with the highest multicollinearity,
#recalculates the vifs, and repeats these same steps until we no longer have
#any vif > 5 in our model
def calculate_vif_(X, threshold):
    dropped=True
    dropped_vars = []
    while dropped:
        dropped=False
        vif = [variance_inflation_factor(X.values, ix) for ix in range(X.shape[1])]

        maxloc = vif.index(max(vif))
        if max(vif) > threshold:
            dropped_vars.append(X.columns[maxloc])
            X = X.drop(X.columns[maxloc], axis = 1)
            dropped=True
    print("Here are the variables that were dropped: ")
    print(dropped_vars)
    return X

predictors = calculate_vif_(predictors, 5.0)
```

Now that we have gotten rid of all the highly correlated variables, we can finally do a linear regression on these predictors. Again, we will attempting to predict city08, or city mpg. The results of the regression are shown below:

```python
#run our regression
x = sm.add_constant(predictors.values)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

Looks like we are almost done! But, there is one more thing we need to take care of. Looking at the p-values of the individual predictors, we notice that a variable  that has a p-value greater than 0.05. We'd like to remove this insignifcant predictor from the model and rerun the regression.

```python
#we'll remove variable 9whose p value's > .05, which means they it is significant
predictors = predictors.drop(predictors.columns[[8]], axis = 1)

#now let's rerun the regression one more time
x = sm.add_constant(predictors.values)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

Now, if we look at the p-values for each predictor again, we see that each of them are significant. That means we are done!

Below we print the final model. We should note that **running this code multiple times will produce slightly different models due to the random sampling imputation we do during the data cleaning step.**

```python
#printing out the model in an equation format

citympg_final_model = ""
param_names = list(predictors)

#add each coefficient and parameter name to equation
for i in range(1, len(result.params)):
    coef = round(result.params[i],3)
    citympg_final_model += (" + " + str(coef) + "(" + param_names[i-1] + ")")
    
citympg_final_model = "city08 = " + str(round(result.params[0],3)) + citympg_final_model
    
print(citympg_final_model)
```

Now that we have this model, what does it mean? Well, looking at the coefficients of the individual predictors we see that by far the variable "electric?" (whether it is an electric car or not) has a very significant influence on city mpg increasing. It's coefficient is 76.478, which means that when all other variables are held constant, a vehicle's city mpg on average goes up about 76.478 mpg if it is an electric car. Other vehicle attributes which increase a car's city mpg according to our model include if the car has diesel fuel and if it has [fuel saving start-stop technology.](https://www.nytimes.com/2016/04/08/automobiles/wheels/start-stop-technology-is-coming-to-cars-like-it-or-not.html?_r=0) Some attributes which negatively affect a vehicle's city mpg more than other include if it has rear-wheel drive or 4-wheel drive.

Let's also do a regression for highway mpg to see if some vehicle attributes affect highway mpg differently than city mpg. We'll follow the same steps we did earlier: first we drop the variables that are unnecessary to keep in our model, remove those with high multicollinearity, and create the model.

```python
#Now let's do one for highway to see if there's any difference between this and city mpg
#following the same steps:

#Creating a linear regression model to predict highway08
#remove unrelated predictors
predictors = df.drop(['year','id', 'make', 'model', 'city08', 'cityA08', 'co2TailpipeGpm',\
                      'co2TailpipeAGpm', 'comb08', 'combA08', 'highway08', 'highwayA08'], \
                     axis = 1).astype(float)

#add a constant term to our regression
x = sm.add_constant(predictors.values)

#convert to np.array
y = df['highway08'].as_matrix()

#run our multicollinear variable remover as before
predictors = calculate_vif_(predictors, 5.0)

#now after getting rid our multicollinearity issue, we rerun our regression
x = sm.add_constant(predictors.values)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

Looking at the p-values of the predictors in this model, it seems we have another insignificant predictor to remove before creating the final model.

```python
#after looking at our results summary, we need to remove two predictors, x5 and x9,
#whose p value > .05, which means they are not significant
predictors = predictors.drop(predictors.columns[[8]], axis = 1)

#now let's rerun the regression one more time
x = sm.add_constant(predictors.values)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

Now everything looks good and we're done! Here are our final models:

```python
#printing out the model in an equation format

highwaympg_final_model = ""
param_names = list(predictors)

#add each coefficient and parameter name to equation
for i in range(1, len(result.params)):
    coef = round(result.params[i],3)

    highwaympg_final_model += (" + " + str(coef) + "(" + param_names[i-1] + ")")
    
highwaympg_final_model = "highway08 = " + str(round(result.params[0],3)) + highwaympg_final_model

print(citympg_final_model, "\n")
print(highwaympg_final_model)
```

For the highway mpg model, we can see as in the city mpg model that whether a car runs on electricity has the most significant influence on highway mpg. Comparing the coeffecients between the models, we see that whether or not a car runs on electricity has a greater impact on gas mileage for city driving than highway driving. ## V: What was the Effect of the 2010 Fuel Economy Regulations?

For our final analyses, we wanted to take a look into how the new regulations passed in 2010 affected the auto industry. Based on our earlier analyses, we could see that it appears that cars' fuel efficiencies only really began increasing after 2010. In order to test this hypothesis, we decided to use a one-sided t-test to see whether the mean fuel efficiency of cars after the regulations was significantly greater than the mean fuel efficiency of cars before the regulations. We'll define the null hypotheses that the mean vehicle gas mileage in both cities and highways before 2011 is equal to the mean vehicle gas mileage in both settings from 2011 onward. Our alternative hypotheses thus state that the mean vehicle gas mileage in both cities and highways before 2011 is less than the mean vehicle gas mileage in both settings afterwards.

```python
#separate the data into 2010 and earlier vs 2011 and later
preRegulation = df.loc[df['year'] <= 2010]
postRegulation = df.loc[df['year'] > 2010]

#t test for city mpg for pre regulation and post regulation
#we need to divide the p-value by two since our test is one-sided
city_t_stat, city_p_val = stats.ttest_ind(preRegulation['city08'],postRegulation['city08'], equal_var=False)
print("City MPG (city08)")
print("T-Statistic: " + str(city_t_stat))
print("P-Value: " + str(city_p_val / 2))
print()

#t test for highway mpg for pre regulation and post regulation
#we need to divide the p-value by two since our test is one-sided
highway_t_stat, highway_p_val = stats.ttest_ind(preRegulation['highway08'],postRegulation['highway08'], equal_var=False)
print("Highway MPG (highway08)")
print("T-Statistic: " + str(highway_t_stat))
print("P-Value: " + str(highway_p_val / 2))
```

Our p-values for both tests are well below 0.05, so we can reject the null hypotheses. It's at least clear that for cars produced after 2011, the average vehicle gas mileage in both cities and highways is significantly higher than those produced beforehand. We can't prove this is a result of the 2010 EPA regulations, but we feel our plots and this statistical test suggest it may be the cause behind this phenomenon.
